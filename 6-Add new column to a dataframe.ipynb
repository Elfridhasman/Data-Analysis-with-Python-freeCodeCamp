## 1.2 Adding a new colum with array

The dataframe has 1000 rows, so we need to create array 1000 elements

```python
# import numpy
import numpy as np
```

```python
# create an array of 1000 elements
language_score = np.arange(0,1000)
```

```python
# length of the array
len(language_score)
```

```python
# adding a new column to datafram with an array
df_exams['language score'] = language_score
df_exams
```

```python
# create random integer numbers between 1 and 100
int_Language_score = np.random.randint(1, 100,1000)
```

```python
# min value inclusive and high value exclusive
min(int_Language_score)
max(int_Language_score)
```

```python
df_exams['int_Language_score'] = int_Language_score
df_exams
```
